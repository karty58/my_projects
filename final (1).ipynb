**INTRODUCTION**:

Human Immunodeficiency Virus (HIV) is a viral infection that attacks the immune system, potentially progressing to Acquired Immunodeficiency Syndrome (AIDS) if left untreated. With no known cure, individuals carry the virus for life, but advancements in medical care, particularly Antiretroviral Therapy (ART), can effectively manage HIV and control its progression. HIV symptoms vary but may include flu-like manifestations and persistent fatigue. As a retrovirus, HIV integrates into the host's DNA, presenting challenges for treatment. AIDS is the advanced stage of HIV, marked by severe immune system compromise and susceptibility to opportunistic infections. While AIDS itself is not transmissible, the HIV virus can be transmitted through various means, emphasizing the importance of preventive measures. This brief overview underscores the significance of timely medical intervention and public health efforts in addressing the impact of HIV/AIDS.

**Loading the data**

```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
```

**DATA 1**

The dataset presents information on the incidence of individuals testing positive for both Tuberculosis (TB) and Human Immunodeficiency Virus (HIV) in various countries from 2007 to 2016. It includes details such as the country's name, country code, the specific year of observation, and the total count of TB patients who tested positive for HIV in each corresponding year. This dataset serves to quantify and highlight the prevalence of co-infection between TB and HIV across different nations, offering valuable insights into trends and patterns over the specified timeframe, and aiding in the formulation of targeted public health strategies.

```python
data1 = pd.read_csv("tb-patients-tested-positive-for-hiv.csv")
# display original
data1
```

**Renaming the columns**

```python
# Rename the column 'TB patients tested positive for HIV - Total' to 'tb_positive_count' in the 'data1' DataFrame
data1 = data1.rename(columns={'TB patients tested positive for HIV - Total': 'tb_positive_count'})
data1# Rename the column 'TB patients tested positive for HIV - Total' to 'tb_positive_count' in the 'data1' DataFrame
def rename_column(dataframe, old_column, new_column):
    """
    Rename a column in a DataFrame.
    (pd.DataFrame, str, str)->str
    """
    dataframe = dataframe.rename(columns={old_column: new_column})
    return dataframe

# Rename the column 'TB patients tested positive for HIV - Total' to 'tb_positive_count'
data1 = rename_column(data1, 'TB patients tested positive for HIV - Total', 'tb_positive_count')
data1
```

**Filtering the rows**

```python
# Filter rows where 'Entity' is 'United States'
data_1 = data1[data1['Entity'] == 'United States']
# display original
data_1
```

**DATA 2**

This dataset offers estimates of Tuberculosis (TB)-related deaths among individuals living with Human Immunodeficiency Virus (HIV) in the United States throughout the period from 2000 to 2016. Specifically focusing on the United States, the dataset includes information such as the country's name (Entity), country code (Code), the respective year of observation, and the estimated number of TB-related deaths among people living with HIV for each year. This dataset provides valuable insights into the dynamics of TB-related mortality within the HIV-positive population over time, serving as a resource for understanding trends and patterns. The information is instrumental in informing public health strategies and interventions targeted at mitigating the impact of TB among individuals living with HIV in the United States.

```python
data2 = pd.read_excel("tb-related-deaths-hiv.xls")
# display original
data2
```

**Filtering rows**

```python
# Filter rows in the 'data2' DataFrame where 'Entity' is 'United States'
data2 = data2[data2['Entity'] == 'United States']
data2
```

**Renaming the Columns**

```python
def rename_column(dataframe, old_column, new_column):
    """
    Rename a column in a DataFrame.
    (pd.DataFrame, int, int)-> int
    """
    dataframe = dataframe.rename(columns={old_column: new_column})
    return dataframe
# Rename the column 'Estimated TB-related deaths among people living with HIV - estimate' to 'TB_death_counts'
data2 = rename_column(data2, 'Estimated TB-related deaths among people living with HIV - estimate', 'TB_death_counts')

# Display the modified DataFrame
data2
```

**Filtering the Rows**

```python
def filter_data_by_year(dataframe, start_year, end_year):
    """
    Filter rows in a DataFrame based on a specified range of years.
    (pd.DataFrame, int, int)-> int
    """
    filtered_data = dataframe[(dataframe['Year'] >= start_year) & (dataframe['Year'] <= end_year)]
    return filtered_data

# Filter rows in data2 for the years 2003 to 2016
data2 = filter_data_by_year(data2, 2003, 2016)

# Display the filtered DataFrame
data2
```

**Merging the Data**

```python
#merging the data_1 and data_2
merged_data = pd.merge(data_1, data2, on=['Entity', 'Code', 'Year'], how='inner')
```

```python
merged_data
```

The merged dataset, obtained through an inner merge of data_1 and data_2 on the 'Entity', 'Code', and 'Year' columns, provides a comprehensive overview of tuberculosis-related data for the United States across multiple years. The resulting DataFrame, named merged_data, includes additional columns such as 'tb_positive_count' and 'TB_death_counts'. Merging datasets is crucial for data enrichment, allowing for the combination of information from different sources to create a more complete and integrated dataset. In this context, the merged data facilitates analysis, enabling a more in-depth examination of the relationship between tuberculosis-positive counts and corresponding death counts over the years. This consolidated dataset is valuable for trend analysis, pattern identification, and gaining insights into how these health-related variables evolve over time for a specific entity.

# **Calculation**

```python
# Calculate and create a new column 'Ratio_TB_pos_to_death' by dividing 'tb_positive_count' by 'TB_death_counts' in 'merged_data'
merged_data['Ratio_TB_pos_to_death'] = (merged_data['tb_positive_count'] - merged_data['TB_death_counts'])
```

```python
# Drop the 'Code' column from the 'merged_data' DataFrame
merged_data = merged_data.drop(columns=['Code'])
merged_data
```
